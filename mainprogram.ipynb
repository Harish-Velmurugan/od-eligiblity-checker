```python
import pandas as pd
import os
import requests
import account as ac
import smtplib
from PyQt5 import QtCore, QtGui, QtWidgets
import tkinter as tk


class Ui_MainWindow(object):
    
    
    def email(self):
        df=pd.read_csv("D:\\project\\a.csv")
        kf=pd.read_csv("D:\\output\\list.csv")
        vdf = df[~ df.batchno.isin(kf.batchno)]
        s=smtplib.SMTP('smtp.gmail.com:587')
        s.ehlo
        s.starttls()
        s.ehlo()
        s.login(ac.email,ac.password)
        msg='Subject: OD result\n\nCongratulations you are eligible for od please wait for the shortlisted list from your HOD\n\n\nWith regards,\npycoders'
        msg1='Subject: OD result\n\nSorry looks like you have not been shortlisted this time please improve your performance and try again, For special events contact your HOD directly\n\n\nWith regards,\npycoders '

        for i in vdf['email_id']:
            s.sendmail(ac.email,i,msg1)
        for i in kf['email_id']:
            s.sendmail(ac.email,i,msg)
        print("Mail Send")
    
    
    def all_requests(self):
        os.startfile("D:\\output\\datar.csv")
  

    def profile(self):
        bt=int(self.lineEdit_2.text())
        df=pd.read_csv("D:\\output\\datar.csv")
        m=tk.Tk()
        l1=tk.Label(m,text='Name = '+df.loc[df['batchno']==bt,'Name'].to_string())
        l2=tk.Label(m,text='Batch Number = '+df.loc[df['batchno']==bt,'batchno'].to_string())
        l3=tk.Label(m,text='CGPA = '+df.loc[df['batchno']==bt,'cgpa'].to_string())
        l4=tk.Label(m,text='DCC Score = '+df.loc[df['batchno']==bt,'dcc'].to_string())
        l5=tk.Label(m,text='IA Average = '+df.loc[df['batchno']==bt,'ia'].to_string())
        l6=tk.Label(m,text='Attendence % = '+df.loc[df['batchno']==bt,'attend'].to_string())
        l7=tk.Label(m,text='Rating = '+df.loc[df['batchno']==bt,'rating'].to_string())
        l1.pack()
        l2.pack()
        l3.pack()
        l4.pack()
        l5.pack()
        l6.pack()
        l7.pack()
        m.mainloop()
    
    
    def calc_rating(self):
        try:
            os.remove('D:\\project\\input.csv')
        except:
            pass    

        #getting the requests from internet

        try:
            r=requests.get('https://docs.google.com/spreadsheets/d/e/2PACX-1vRO1MatPj2ohD37NyowtQHkY7t4mj4_KlB_Kun_Ixma3K6eD58Ua_ijftfmkwvitFSjBqtG2VtvbdnC/pub?output=csv')
        except:
            print("Internet is not available")
        
        with open('D:\\project\\input.csv','wb') as f:
            f.write(r.content)

        #reading the names of the persons who have submitted the requests

        idf=pd.read_csv("D:\\project\\input.csv")

        #reading the database of records

        cdf=pd.read_csv("D:\\project\\data.csv")

        #filtering the datas of the persons who have submitted thier request

        ddf=cdf[cdf.batchno.isin(idf.Batch_Number)]

        #removing previous files

        try:
            os.remove('D:\\project\\a.csv')
        except:
            pass

        #creating a file with filtered datas

        ddf.to_csv("D:\\project\\a.csv",index=None)

        #reading the datas of persons who submitted the request from filtered data

        df=pd.read_csv("D:\\project\\a.csv")

        #assigning rating to be 0 initially

        df["rating"]=0

        #calculating ratings for all persons 


        df.loc[df["attend"]>90,"rating"]+=.5
        df.loc[df["cgpa"]>8,"rating"]+=1
        
        
        df.loc[df["ia"]>60,"rating"]+=1
        df.loc[df["nomakp"]=="yes","rating"]+=1
        df.loc[df["clrmkp"]=="yes","rating"]+=.5
        df.loc[df["dcc"]>=400,"rating"]+=2

        #filtering persons with rating greater than or equal to 3

        kf=df.loc[df["rating"]>=3].sort_values(by='rating',ascending=False)
        df.to_csv("D:\\output\\datar.csv",index=None)
        #removing previous files

        try:
            os.remove("D:\\output\\list.csv")
        except:
            pass

        #creating the final list with eligible people

        kf.to_csv("D:\\output\\list.csv",index=None)
        print("Sucess in generating file")
        
    def eligible(self):
        os.startfile("D:\\output\\list.csv")
    
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1078, 823)
        MainWindow.setStyleSheet("*{\n"
"    background-image: url(:/download/download.jpg);\n"
"   background:rgb(255,255,255);\n"
"   font-size:1Spx;\n"
"   font-family:Century gothic ,sans-serif;\n"
"}\n"
"QFrame{\n"
"border :Solid 10px rgba(0,0,0);\n"
"background-image:url(C:\\Users\\ASUS\\Downloads\\wallpapers\\download.jpg)\n"
"}\n"
"")
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.frame = QtWidgets.QFrame(self.centralwidget)
        self.frame.setGeometry(QtCore.QRect(0, 0, 1101, 781))
        font = QtGui.QFont()
        font.setFamily("Century gothic")
        font.setPointSize(18)
        self.frame.setFont(font)
        self.frame.setStyleSheet("background-image: url(:/download/download.jpg);")
        self.frame.setFrameShape(QtWidgets.QFrame.StyledPanel)
        self.frame.setFrameShadow(QtWidgets.QFrame.Raised)
        self.frame.setObjectName("frame")
        self.pushButton = QtWidgets.QPushButton(self.frame)
        self.pushButton.setGeometry(QtCore.QRect(160, 280, 161, 28))
        self.pushButton.setStyleSheet("font: 9pt \"MS Shell Dlg 2\";\n"
"color: rgb(255, 255, 255);")
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.profile)
        self.lineEdit = QtWidgets.QLineEdit(self.frame)
        self.lineEdit.setGeometry(QtCore.QRect(160, 190, 161, 22))
        self.lineEdit.setStyleSheet("color: rgb(255, 255, 255);")
        self.lineEdit.setText("")
        self.lineEdit.setEchoMode(QtWidgets.QLineEdit.Normal)
        self.lineEdit.setObjectName("lineEdit")
        self.lineEdit_2 = QtWidgets.QLineEdit(self.frame)
        self.lineEdit_2.setGeometry(QtCore.QRect(160, 220, 161, 22))
        self.lineEdit_2.setStyleSheet("color: rgb(255, 255, 255);")
        self.lineEdit_2.setObjectName("lineEdit_2")
        self.label = QtWidgets.QLabel(self.frame)
        self.label.setGeometry(QtCore.QRect(30, 70, 55, 16))
        self.label.setObjectName("label")
        self.label_2 = QtWidgets.QLabel(self.frame)
        self.label_2.setGeometry(QtCore.QRect(30, 90, 281, 51))
        font = QtGui.QFont()
        font.setFamily("Agency FB")
        font.setPointSize(36)
        font.setBold(False)
        font.setItalic(False)
        font.setWeight(50)
        self.label_2.setFont(font)
        self.label_2.setStyleSheet("color: rgb(255, 255, 255);\n"
"font: 75 8pt \"MS Shell Dlg 2\";\n"
"font: 36pt \"Agency FB\";")
        self.label_2.setObjectName("label_2")
        self.label_3 = QtWidgets.QLabel(self.frame)
        self.label_3.setGeometry(QtCore.QRect(840, 650, 231, 61))
        font = QtGui.QFont()
        font.setFamily("Agency FB")
        font.setPointSize(26)
        font.setBold(False)
        font.setItalic(False)
        font.setWeight(50)
        self.label_3.setFont(font)
        self.label_3.setStyleSheet("color: rgb(255, 255, 255);\n"
"font: 26pt \"Agency FB\";")
        self.label_3.setObjectName("label_3")
        self.pushButton_2 = QtWidgets.QPushButton(self.frame)
        self.pushButton_2.setGeometry(QtCore.QRect(860, 190, 171, 41))
        font = QtGui.QFont()
        font.setFamily("Agency FB")
        font.setPointSize(20)
        font.setBold(False)
        font.setItalic(False)
        font.setWeight(50)
        self.pushButton_2.setFont(font)
        self.pushButton_2.setStyleSheet("font: 20pt \"Agency FB\";\n"
"color: rgb(255, 255, 255);")
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.all_requests)
        self.pushButton_3 = QtWidgets.QPushButton(self.frame)
        self.pushButton_3.setGeometry(QtCore.QRect(860, 290, 171, 51))
        self.pushButton_3.setStyleSheet("color: rgb(255, 255, 255);\n"
"font: 19pt \"Agency FB\";")
        self.pushButton_3.setObjectName("pushButton_3")
        self.pushButton_3.clicked.connect(self.email)
        self.pushButton_4 = QtWidgets.QPushButton(self.frame)
        self.pushButton_4.setGeometry(QtCore.QRect(860, 130, 171, 41))
        font = QtGui.QFont()
        font.setFamily("Agency FB")
        font.setPointSize(20)
        font.setBold(False)
        font.setItalic(False)
        font.setWeight(50)
        self.pushButton_4.setFont(font)
        self.pushButton_4.setStyleSheet("font: 20pt \"Agency FB\";\n"
"color: rgb(255, 255, 255);")
        self.pushButton_4.setObjectName("pushButton_4")
        self.pushButton_4.clicked.connect(self.calc_rating)
        self.pushButton_5 = QtWidgets.QPushButton(self.frame)
        self.pushButton_5.setGeometry(QtCore.QRect(860, 240, 171, 41))
        font = QtGui.QFont()
        font.setFamily("Agency FB")
        font.setPointSize(18)
        font.setBold(False)
        font.setItalic(False)
        font.setWeight(50)
        self.pushButton_5.setFont(font)
        self.pushButton_5.setStyleSheet("font: 18pt \"Agency FB\";\n"
"color: rgb(255, 255, 255);")
        self.pushButton_5.setObjectName("pushButton_5")
        self.pushButton_5.clicked.connect(self.eligible)
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 1078, 23))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "OD ELIGIBILITY CHECKER "))
        self.pushButton.setText(_translate("MainWindow", "CHECK"))
        self.pushButton.setShortcut(_translate("MainWindow", "Return"))
        self.lineEdit.setPlaceholderText(_translate("MainWindow", "Username"))

        self.lineEdit_2.setPlaceholderText(_translate("MainWindow", "Batch no"))
        self.label.setText(_translate("MainWindow", "Name : "))
        self.label_2.setText(_translate("MainWindow", "Check Profile"))
        self.label_3.setText(_translate("MainWindow", "#PyCoders"))
        self.pushButton_2.setText(_translate("MainWindow", "Show Full List"))
        self.pushButton_3.setText(_translate("MainWindow", "Send Email"))
        self.pushButton_4.setText(_translate("MainWindow", "Generate List"))
        self.pushButton_5.setText(_translate("MainWindow", "Show Eligible List"))

import image_rc

if __name__ == "__main__":
    try:
        import sys
        app = QtWidgets.QApplication(sys.argv)
        MainWindow = QtWidgets.QMainWindow()
        ui = Ui_MainWindow()
        ui.setupUi(MainWindow)
        MainWindow.show()
        sys.exit(app.exec_())
    except:
        exit()
```
